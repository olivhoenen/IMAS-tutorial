# Welcome to the IMAS tutorial!

Welcome to the IMAS (Integrated Modelling and Analysis Suite) Tutorial!
This series of notebooks will introduce you to IMAS and various of its software components.
Note that this tutorial is a living document and covers only open-source software compoments.
New sections will be added regularly when additional components are ready and made openly accessible and to describe additional use cases.

This tutorial is designed for newcomers to IMAS who should follow topics in order.
This material can also allow be useful for long-time users who may want to refresh their knowledge or jump directly to a topic and discover the associated recommended tools.

## Content of the tutorial

Click on the links below to open each notebook:

### 1. [Introduction to IMAS](01_Introduction_to_IMAS.ipynb) 
Learn what IMAS is and understand its key concepts and terminology.
- What is IMAS?
- Architecture and concepts
- Some typical use cases


### 2. [IMAS Data Dictionary](02_IMAS_Data_Dictionary.ipynb) 
Explore the structure and organization of IMAS data model.
- Key concepts for the IMAS data model
- IMAS-Data-Dictionary project and IDS structures
- Common sub-structures
- Install and use the IMAS-Data-Dictionary


### 3. Data Access 
Learn how access to work with IMAS IDS data using different programming languages.

#### 3.0 [IMAS-Core](03_IMAS_Core.ipynb)
Most users won't directly call this low-level library, but it may be useful to learn some key concepts and terminology.
- Key concepts for the IMAS data access libraries
- IMAS-Core project and the notion of backends

Extra topics will be covered in the specific language support sections

#### 3.1 [IMAS-Python](03_IMAS_Python.ipynb)
The recommended library to interact with IDS in Python.
- Install and use the library
- Read and write IDS data
- Interact with time: timebases and slicing 

#### 3.2 IMAS-Fortran *coming soon...*
The recommended library to interact with IDS in Fortran.


### 4. Data Exploration and Visualization
Various tools to explore the content of IMAS data.

#### 4.1 [IDStools](04_IDStools.ipynb)
A simple and command-line-oriented toolbox to explore, plot and manipulate IDS data. 
- Install and use the package
- Explore *data-entries*
- Ad-hoc plot examples 
- Manipulate IDS

#### 4.2 IBEX *coming soon...*
A graphical tool to browse, plot (profiles in 1D and 2D maps) and compare any IDS quantities.

#### 4.3 IMAS-ParaView *coming soon...*
An IMAS data plugin for ParaView, for complex data visualization pipelines in 2D and 3D. 


### 5 [Simulation database management](05_SimDB.ipynb) 
Discover the SimDB solution to assemble and manage databases of IMAS simulations.
- Install and use the command line client
- Configure a server (e.g. ITER scenarios DB, requires ITER user account)
- Search and download a simulation
- Ingest and upload a simulation
