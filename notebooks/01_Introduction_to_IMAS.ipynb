# Introduction to IMAS

In this part of the tutorial you will learn IMAS stands for, how is it designed and what it is made for.

## What is IMAS?

IMAS stands for the Integrated Modelling & Analysis Suite, it's the collection of physics software to support ITER operations and research. It is composed of a set of plasma modelling tools for the systematic planning (evaluation of candidate plasma operating scenarios and assistance in the development of plasma control strategies) and the analysis of each ITER discharge.

To reach its numerous ambitious goals, IMAS relies on expertise and codes from close collaborators within ITER Members (CN-EU-IN-JP-KR-RF-US), as well as experimental data and computing facilities within their respective domestic programs. The chosen strategy to aim at a seamless inclusion of interoperable modelling and analysis software is to decompose IMAS in three nested layers:

- a standardized **data model**;
- an extensible set of **generic software tools**;
- a portfolio of interoperable physics **applications**.


## Key Concepts

### Data model

- Covers all diagnostic systems, actuators and the relevant plant systems to allow for the most comprehensive description of the plasma behaviour.
- Is machine agnostic, such as it can describe and be applied to existing experimental device (at least for validation purposes).
- Defines data structures that can serve as coupling interfaces to build integrated modelling simulation and analysis workflows.

### Generic software tools

- Comply with the data model.
- Are not mandatory but shall help developers while interacting with the data model without reinventing the wheel.
- Cover topics such as data access, data visualization, code coupling and database management.

### Applications

- Can be written in various programming languages.
- Shall as much as possible use machine description expressed within the data model and avoid assumed parameters.
- Existing ones can be adapted to IMAS by the addition of a thin wrapper layer that converts its internal data structures to IMAS data model.


## Some use cases

- Integrated modeling workflows combining multiple physics codes
- Experimental data analysis and plasma reconstruction chains
- Comparing experimental measurements with simulation predictions
- Long-term data archival and retrieval (data + metadata)
- Collaborative research across institutions
