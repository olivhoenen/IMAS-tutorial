# IMAS Data Dictionary

The [IMAS-Data-Dictionary](https://github.com/iterorganization/IMAS-Data-Dictionary) (often referred to as the _DD_) is the foundation of the IMAS ecosystem. It defines the structure, semantics, and relationships of all data used in IMAS.


## Concept and terminology

This data model is driven by the idea of becoming a common standard for describing any kind of relevant data that supports IMAS goals. 
It has the following primary design features:

- It is **machine agnostic**, as it shall be applicable to ITER but also to other tokamaks such as applications can be validated on existing devices;
- It can cover both **simulation** and **experimental** data in order to meet IMAS goals to support the design and preparation of a pulse, as well as its analysis after the acquisition (possibly at different sampling rate) of various signals from sensors and diagnostics installed on the machine;
- It is organized as an **ontology of data structures** that have a narrow and specific scope (e.g. a type of diagnostic, a type of actuator, a type physics process or phenomena, a region of interest, etc...) but are composable to propose a larger and more complete description;
- It needs to be **extensible**, to allow the incorporation of new systems and ideas from contributors, and needs to be carefully version-controlled such as the same exact version can be used as a common _interface_ between software developed by different groups in order to ease exchange of information. By consequence, the data structures that compose the ontology are named **Interface Data Structures (IDS)**;
- Each IDS (the *root* of the tree) has a hierarchical representation made of several nested *nodes* (**structures** and **array of structures**) that consist of metadata only and are used to group/organize topics and sub-systems, and *leaves* (**fields**) that actually stores data in the supported native types (integer, floating points, characters or complex data) and various dimensionality (from 0D to 6D).


## Online documentation

The main outcome of the IMAS-Data-Dictionary is the description and metatada associated with each node and leaf of each IDS. This can be consulted online for either the [latest](https://imas-data-dictionary.readthedocs.io/en/latest/) development version or the most recent tagged releases [3.41](https://imas-data-dictionary.readthedocs.io/en/3.41.0-post/), [3.42](https://imas-data-dictionary.readthedocs.io/en/3.42.2-migration/), [4.1](https://imas-data-dictionary.readthedocs.io/en/4.1.1/).

## IMAS-Data-Dictionary repository 

The [IMAS-Data-Dictionary repository](https://github.com/iterorganization/IMAS-Data-Dictionary) contains the implementation of the data model as a set of schemas that are then assembled to generate the hierarchical description of all IDSs in XML format. 

You may want to install locally a given version of the IMAS-Data-Dictionary, which is made available as a [Python package on PyPI](https://pypi.org/project/imas-data-dictionary/) for your conveniency. Will bring the following:

- all XML files that describes the data model
   - `data_dictionary.xml` defines the structures, field names, units and coordinates that comprise all IDSs;
   - `*_identifier.xml` describe `index, name, description` values for various set of [_identifiers_](https://imas-data-dictionary.readthedocs.io/en/latest/reference_ids.html) that provide enumerated list for various component of an IDS (e.g. grid types,m coordinate systems, particle species, heating systems, ...);
- simple command line utilities for your convenience: `dd_doc` and `idsinfo`.

### Alternate package with all versions 

For applications that needs to have a description of all versions of the DD (e.g. [IMAS-Python](03_IMAS_Python.ipynb) which use this knowledge to perform some level of data conversion between versions), a complementary package called `imas-data-dictionaries` is automatically published on PyPI when a new version of `imas-data-dictionary` is released. This alternate package only contains XML resources for each version since 3.22.0 (released in 2019!).
