# IMAS Core 

[IMAS-Core](https://github.com/iterorganization/IMAS-Core) (formerly known as the _Access-Layer Low-Level_) is a low-level C++ library that abstracts the details of the data storage on disk (or different data sources) from the end-user IDS API like [IMAS-Python](03_IMAS_Python.ipynb), IMAS-Fortran, IMAS-MATLAB, IMAS-Cpp and IMAS-Java (previously referred to as the _Access-Layer High-Level Interfaces_).

While most users may never directly call the IMAS-Core API, it is useful to know about some of its concepts and to learn how to configure and install it.


## Concept and terminology

Abstracting data storage details from the API used by users in their code allows to easily switch from a given I/O store to another, giving more flexibility without complexifying the application. IMAS-Core provide the following key elements:

- IMAS data is organized in a [**data entry**](https://imas-data-dictionary.readthedocs.io/en/latest/dataentry_and_occurrence.html) that comprises one or more [IDS](02_IMAS_Data_Dictionary.ipynb) types to describe a simulation results or a pulse; 
- A C-based API for portability to other programming languages;
- The API consist in **context management** functions (start operation at a _data entry_, _IDS_ or _array of structure_ levels) and functions for actual read/write operations (reading and writing all the supported data types: integers, floating points, characters, complex);
- The data entry API relies on a [**URI syntax**](https://imas-core.readthedocs.io/en/latest/user_guide/uris_guide.html) to ensure flexibility and extensibility;
- A Cython-based binding of the API for Python;
- Several (optional) **backends** with their own [specificities](https://imas-core.readthedocs.io/en/latest/user_guide/backends_guide.html#backend-comparison): `hdf5`, `mdsplus`, `uda`, `ascii`, `memory`, `flexbuffers`.

<div class="alert alert-block alert-info">
<b>Info:</b> The IMAS-Core is generally not aware of the internals of IDS structure's details, so it does not depend on IMAS-Data-Dictionary. There are two exceptions: at build time when using the <code>mdsplus</code> backend which needs <i>models</i> to be generated for a given version of the DD, and at runtime when using the <code>uda</code> backend with its default <a href=https://imas-core.readthedocs.io/en/latest/user_guide/backends_guide.html#query-keys-specific-for-the-uda-backend><code>cache_mode=ids</code></a> option.</div>



### Configuration and install

Building from sources with CMake offers easy control on all [configuration](https://imas-core.readthedocs.io/en/latest/user_guide/installation.html) options and let you decide which optional backends (and their respective dependencies) and bindings (Python) you want to support. 

For easier startup with Python applications, IMAS-Core can now be installed from sources with `pip`. Note that it is using [scikit-build](https://scikit-build.readthedocs.io/en/latest/) to interact with CMake, so if you don't want the default configuration options you will need to pass CMake options via pip with the `-C skbuild.cmake.define.OPTION=VALUE` [syntax](https://imas-core.readthedocs.io/en/latest/user_guide/installation.html#install-from-source-using-pip).

But it can get even simpler than that as an `imas-core` Python package is also available on PyPI for Linux, MacOSX (arm) and Windows. 

### Install of the package

```python
!pip install imas-core
import imas_core
print(f"installed version of IMAS-Core = {imas_core.__version__}")
```

### Simple test with HDF5 backend

This test allow us to take a look at how files are organized in a data entry. First, all files related to a given data entry are stored inside the same folder. In case of the `hdf5` backend, you will find in this folder a `master.h5` which will contains one or several links to individual IDS files named as `<idsname>.h5` (e.g. `equilibrium.h5`, ` wall.h5`, `pf_active.h5`, ...) and `<idsname>_<occurrence>.h5` in case you have several occurrences of the same IDS type (e.g. `equilibrium_1.h5`). 


```python
# now creating a new data-entry using the HDF5 backend (don't mind the API here, you will not see this as an end-user)
imas_core._al_lowlevel.al_begin_dataentry_action("imas:hdf5?path=./testh5",imas_core.al_defs.CREATE_PULSE)
!ls -l ./testh5
```

### Test with MDSplus backend

Other backends will use different files and names but these files will still belong the a unique folder. E.g. with the `mdsplus` backend, the folder will contain always (and only) three files: `ids_001.characteristics`, `ids_001.tree` (these are set by the models) and `ids_001.datafile` that contains the data values.

Note that the <a href=https://imas-core.readthedocs.io/en/latest/user_guide/configuration.html#uda-client-configuration-to-reach-the-server-at-iter><code>uda</code> backend</a> is also present in the Linux package but not yet for MacOSX or Windows.</div>


```python
# verifying that indeed the mdsplus backend is not part of this package, so it won't create the targeted folder for the data entry!
imas_core._al_lowlevel.al_begin_dataentry_action("imas:mdsplus?path=./testmds",imas_core.al_defs.CREATE_PULSE)
!ls -l ./testmds
```
