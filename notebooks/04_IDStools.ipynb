# IDStools

[IDStools](https://github.com/iterorganization/IDStools) is a toolbox with many little Python scripts to help your plot, explore and manipulate data entries and IDS data from the command line.


## Concept 

All the scripts inside this toolbox are relying on [IMAS-Python](03_IMAS_Python.ipynb) to interact with IMAS IDS data and are typically contributed over years from a variety of users. The source code was reorganized to better isolate functionalities from the CLI scripts, so you can also import and reuse them in your own applications and not only via the command line.

Tools can be categorized in three categories:

- [analysis tools](https://imas-idstools.readthedocs.io/en/latest/analysis_tools.html) are focusing on plotting data coming from one or several IDSs; 
- [IDS manipulation tools](https://imas-idstools.readthedocs.io/en/latest/ids_manipulation_tools.html) are focusing on exploration and simple operations on the IDS data;
- [database tools](https://imas-idstools.readthedocs.io/en/latest/database_tools.html) are there for historical reasons, can still be used but are mostly deprecated and to be replaced by [SimDB](05_SimDB.ipynb).

Arguments of the scripts make extensive use of the [IMAS URI syntax](https://imas-data-dictionary.readthedocs.io/en/latest/IMAS-URI-scheme.html), including its [fragment's specification](https://imas-data-dictionary.readthedocs.io/en/latest/IMAS-URI-scheme.html#fragment).

## Explore and manipulate scripts

### [`idslist`](https://imas-idstools.readthedocs.io/en/latest/idslist.html) to list IDSs and times present in a data entry

```python
!idslist -u iter_scenario_53298_seq1_DD4.nc 
```

### [`idsprint`](https://imas-idstools.readthedocs.io/en/latest/idsprint.html) to print the content of an IDS 

```python
!idsprint -u iter_scenario_53298_seq1_DD4.nc#summary/global_quantities
```

### `eqdsk2ids` to convert a geqdsk file to IDS

```python
# let's download an open access geqdsk file, courtesy of SPARCPublic
!wget https://raw.githubusercontent.com/cfs-energy/SPARCPublic/refs/heads/main/EQlibrary/sparc_1358.EQDSK_PsiOver2pi_negIp_negBt_negFpol -O sparc.EQDSK

# here we need to specify the input COCOS for the eqdsk file (FreeGS uses 3 or 7), while the output is fixed by IMAS DD
!eqdsk2ids -g sparc.EQDSK -d imas:hdf5?path=./fromeqdsk -c 3
```
