# SimDB

[SimDB](https://github.com/iterorganization/SimDB) is a client/server tool that helps managing IMAS simulations. Here we will cover the light client usage to query a simulation database from a remote server (this example will use the ITER scenarii server, access to which requires authentication with an ITER account).


## Concept and terminology

The SimDB solution is composed of several components: the server, the command-line client, and a web UI client (the dashboard).

The **server** builds and stores a database of metadata that describes scenario's key-parameters. It also stores all the data associated with the simulation: input files (data and code-specific parameters) for reproducibility and output data (possibly with code-specific restart files to allow extending the simulation). While it can be used outside of IMAS in principle, IMAS DD brings standardized naming and definition for metadata, which improves queries and comparisons between simulations. It also allows automatation of metadata extraction, all you need is to have an IMAS data entry as part of your simulation outputs, and this data entry needs to contain a `summary` IDS (the metada schema is currently based on this IDS). 

The **command-line client** can be used for several purposes: 

- a simulation data providers can describe (in a _manifest_) and register (_ingest_) simulations into a local _simulation_ database;
- in conjunction with a server (a _remote_), local clients can upload simulations (_push_ metadata and potentially data depending on the server configuration) from their local database into the remote server for sharing them with others. A user can also download (_pull_) a simulation from a remote into the local simulation database;
- the client can also be used to _list_, obtain _info_ and _query_ a simulation database (local or remote) to identify simulations of interest.

A web client is also available (e.g. an instance is running for the ITER scenarios database at [simdb.iter.org](https://simdb.iter.org/)) and allows for searching and comparing (to some extent) simulations from the catalog.

## Set a [`remote`](https://simdb.readthedocs.io/en/latest/user_guide.html#remote-simdb-servers) for your client 

You can directly use the client to register ([`ingest`](https://simdb.readthedocs.io/en/stable/tutorial.html#ingesting-the-manifest)) a new simulation locally, but here as we don't have a simulation yet, we will use the client to query and find a simulation of interest from the [ITER remote server](https://simdb.readthedocs.io/en/stable/iter_remotes.html).


```python
# here we configure a remote pointing to the ITER scenario DB server, which is being authentication
# we also make it the default remote so we don't have to specify it in follow-up remote commands
!simdb remote config new ITER https://simdb.iter.org/scenarios/api/ --default --firewall F5 

# you can also configure your username on the server, so you want to ask to enter it 
#!simdb remote config set-option ITER username <your-ITER-username>

# check first if we have remotes
!simdb remote config list
```

## Find simulations with `list` and `query`

```python
# list all simulations (returns their alias or UUID, and the name of the simulation code that produced it)
!simdb remote list -m code.name
```

```python
# you can get more details on a given simulation by giving the simulation UUID or alias to the info command
!simdb remote info 134173/106
```

```python
# and you can search matching simulations against criteria that corresponds quantities inside the summary IDS
!simdb remote query global_quantities.b0.value=agt:-3 composition.neon.value=ge:0.02 -m code.name -l 15
```

## Download simulation metadata and data with `pull`

```python
# check first what you have in your local simulation DB
!simdb simulation list
```

```python
# once you've found a simulation of interest, you can pull it in your local simulation DB for further analysis or restart
!simdb simulation pull kims12/jetto/iter/53301/jan1725/seq-1 ./sim_folder
```

```python
# now the simulation is described in your local DB
!simdb simulation list
```

```python
# and you have all inputs and outputs for this simulation in the targeted folder
!ls -l ./sim_folder
```
